### One Size Images

```python
def copy_images_with_size(source_folder, destination_folder, target_width, target_height):
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    for filename in os.listdir(source_folder):
        file_path = os.path.join(source_folder, filename)
        if os.path.isfile(file_path):
            # read image
            image = cv2.imread(file_path)
            
            # Get the image dimensions
            width = image.shape[1]
            height = image.shape[0]

            if width == target_width and height == target_height:
                # Copy the image to the destination folder
                shutil.copy(file_path, destination_folder)
                print(f"Copied {filename} to {destination_folder}")
```

### Resize Image (Factor)

```python
def resize_images_factor(input_dir, output_dir, downscale_factor):
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Iterate through the subdirectories in the input directory
    for root, dirs, files in os.walk(input_dir):
        for dir in dirs:
            # Create corresponding subdirectories in the output directory
            subdir_path = os.path.join(output_dir, root, dir)
            if not os.path.exists(subdir_path):
                os.makedirs(subdir_path)

        for file in files:
            # Read the image
            image_path = os.path.join(root, file)
            image = cv2.imread(image_path)

            # Downsample the image to the factor of x
            # Calculate the target size for downsampling
            target_width = image.shape[1] // downscale_factor
            target_height = image.shape[0] // downscale_factor

            # Resize the image using cv2.resize with bicubic interpolation
            downsampled_image = cv2.resize(image, (target_width, target_height), interpolation=cv2.INTER_CUBIC)

            # Save the resized image to the output directory
            output_path = os.path.join(output_dir, root, file)
            cv2.imwrite(output_path, downsampled_image)

    print("Image transformation completed!")
```

```python
resize_images_factor(input_directory, output_directory, downscale_factor)
```

### Resize Image (Fixed)

```python
def resize_images_fixed(input_dir, output_dir, target_width, target_height):
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Iterate through the subdirectories in the input directory
    for root, dirs, files in os.walk(input_dir):
        for dir in dirs:
            # Create corresponding subdirectories in the output directory
            subdir_path = os.path.join(output_dir, root, dir)
            if not os.path.exists(subdir_path):
                os.makedirs(subdir_path)

        for file in files:
            # Read the image
            image_path = os.path.join(root, file)
            image = cv2.imread(image_path)

            # Downsample the image to the factor of x
            # Calculate the target size for downsampling
            target_width = target_width
            target_height = target_height

            # Resize the image using cv2.resize with bicubic interpolation
            downsampled_image = cv2.resize(image, (target_width, target_height), interpolation=cv2.INTER_CUBIC)

            # Save the resized image to the output directory
            output_path = os.path.join(output_dir, root, file)
            cv2.imwrite(output_path, downsampled_image)

    print("Image transformation completed!")
```

### Gather Image Into One Folder and Renamed it

```python
def gather_and_rename_images(input_dir, output_dir):
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    # Iterate through the subdirectories in the input directory
    for root, dirs, files in os.walk(input_dir):
        for file in files:
            # Get the category and subcategory from the directory structure
            category = os.path.basename(root)
            subcategory = os.path.basename(os.path.dirname(root))

            # Construct the new filename in the desired format
            filename, extension = os.path.splitext(file)
            
            # Split the filename into parts using regex
            pattern = r'([A-Za-z]+)(\d+)'

            # Extract the class and ID from the filename
            match = re.match(pattern, filename)

            file_class = match.group(1)
            imageId = match.group(2)
            
            new_filename = f"{subcategory}{imageId}_{category}{extension}"

            # Generate the source and destination paths
            source_path = os.path.join(root, file)
            destination_path = os.path.join(output_dir, new_filename)

            # Copy and rename the image file to the output directory
            shutil.copyfile(source_path, destination_path)
            
    print("Image gathering and renaming completed!")
```

```python
# Set the paths for the input directory and output directory
input_directory = "datav3"
output_directory = "datav3_preprocessed"
```

```python
# Call the function to gather and rename the images
gather_and_rename_images(input_directory, output_directory)
```

### Split One Folder to Three Folders with Sub Folders Class

```python
def split_images(source_folder, destination_folder):
    # Create the train, validation, and testing folders
    train_folder = os.path.join(destination_folder, 'train')
    validation_folder = os.path.join(destination_folder, 'validation')
    testing_folder = os.path.join(destination_folder, 'testing')

    os.makedirs(os.path.join(train_folder, 'Accident'))
    os.makedirs(os.path.join(train_folder, 'NonAccident'))
    os.makedirs(os.path.join(validation_folder, 'Accident'))
    os.makedirs(os.path.join(validation_folder, 'NonAccident'))
    os.makedirs(os.path.join(testing_folder, 'Accident'))
    os.makedirs(os.path.join(testing_folder, 'NonAccident'))

    # Get the list of all image files in the source folder
    image_files = [f for f in os.listdir(source_folder) if os.path.isfile(os.path.join(source_folder, f))]

    # Shuffle the image files randomly
    random.shuffle(image_files)

    # Calculate the number of images for each split
    total_images = len(image_files)
    train_count = math.ceil(total_images * 0.8)
    validation_count = math.ceil(total_images * 0.1)
    test_count = total_images - train_count - validation_count

    # Move/Copy the images to their respective folders based on the split
    for i, image_file in enumerate(image_files):
        #image_class = 'Accident' if 'Accident' in image_file else 'NonAccident'
        
        # Split the extension and filename
        name = image_file.split('.')[0]
        
        # Split the filename into parts using regex
        pattern = r'([A-Za-z]+)(\d+)'
        
        # Extract the class and ID from the filename
        match = re.match(pattern, image_file)
        image_class = match.group(1)
        
        src_path = os.path.join(source_folder, image_file)

        if i < train_count:
            dst_path = os.path.join(train_folder, image_class, image_file)
        elif i < train_count + validation_count:
            dst_path = os.path.join(validation_folder, image_class, image_file)
        else:
            dst_path = os.path.join(testing_folder, image_class, image_file)

        shutil.copy(src_path, dst_path)

    print('Image splitting completed successfully.')
```

### Split One Folder Images to Two Folders

```python
def copy_and_rename_files(source_directory, destination_directory_accident, destination_directory_non_accident):
    # Initialize counters for renaming files
    accident_counter = 1
    non_accident_counter = 1
    
    # Create destination directories if they don't exist
    os.makedirs(destination_directory_accident, exist_ok=True)
    os.makedirs(destination_directory_non_accident, exist_ok=True)

    # Iterate over all files in the source directory
    for filename in os.listdir(source_directory):
        # Split the extension and filename
        name = filename.split('.')[0]
        
        # Split the filename into parts using regex
        pattern = r'([A-Za-z]+)(\d+)'
        
        # Extract the class and ID from the filename
        match = re.match(pattern, filename)
        
        file_class = match.group(1)
        file_id = match.group(2)
        
        # Create the new filename based on the class and counter
        new_filename = ""
        if file_class == "Accident":
            new_filename = "Accident" + str(accident_counter) + '.jpg'
            accident_counter += 1
            destination = destination_directory_accident
        elif file_class == "NonAccident":
            new_filename = "NonAccident" + str(non_accident_counter) + '.jpg'
            non_accident_counter += 1
            destination = destination_directory_non_accident
        
        # Copy the file to the appropriate destination directory with the new filename
        shutil.copy(os.path.join(source_directory, filename), os.path.join(destination, new_filename))

    print("File copying and renaming completed.")
```

```python
source_directory = "./datav3/"
destination_directory_accident = "./datav3_splitted/Accident/"
destination_directory_non_accident = "./datav3_splitted/NonAccident/"
```

```python
copy_and_rename_files(source_directory, destination_directory_accident, destination_directory_non_accident)
```

### Split One Folder Images to Train,Val, Test by Filename Format

```python
def split_by_filename(source_folder, destination_folder):
    # Create the destination folder if it doesn't exist
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # Iterate through the files in the source folder
    for filename in os.listdir(source_folder):
        if os.path.isfile(os.path.join(source_folder, filename)):
            # Parse the category, and subcategory from the filename
            category, subcategory = filename.split('_')
            
            # Split the category and item id
            match = re.match(r"([a-zA-Z]+)(\d+)", category) 
            
            if match:
                category = match.group(1) # Output: train
                identifier = match.group(2) # Output: 12
                
            else:
                print(f"Invalid filename format: {category}")
                

            # Create the destination subfolder path
            subfolder_path = os.path.join(destination_folder, category, subcategory)

            # Create the subfolder if it doesn't exist
            if not os.path.exists(subfolder_path):
                os.makedirs(subfolder_path)

            # Copy the file to the destination subfolder
            source_path = os.path.join(source_folder, filename)
            destination_path = os.path.join(subfolder_path, filename)
            shutil.copyfile(source_path, destination_path)

    print("File copying completed!")
```
